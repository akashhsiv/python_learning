### tuples

```python
#tuples are ordered, unchangeable,has ()paranthesis, allow duplicates

tpl=('a','k','q','s','p')
print(tpl[2])  #q
print(tpl[-1])  #p
print(tpl[:4]) #'a','k','q','s'
print(tpl[1:5])  #'k','q','s','p'
print(tpl[-4:])   #'k','q','s','p'


if 'a' in tpl:    #yes
    print('yes')
else:
    print('no')    

#UPDATE TUPLES
#once tuple is created you can't change its value    tuple is unchangeable

#but there is way convert your tuple to list update it and convert it to tuple again

ntpl=list(tpl)  #['a', 'k', 'q', 's', 'p']
print(ntpl)

ntpl.insert(1,'e')   #adds in to list    #['a', 'e', 'k', 'q', 's', 'p']

print(ntpl)

#another method        removes the current item and adds new one
ntpl[1]='r'   #['a', 'r', 'k', 'q', 's', 'p']
print(ntpl)


#add item in tuple
y=ntpl
print(y)  #['a', 'r', 'k', 'q', 's', 'p']

y.append(5)   #['a', 'r', 'k', 'q', 's', 'p', 5]
print(y)

s=tuple(y)   #('a', 'r', 'k', 'q', 's', 'p', 5)
print(s)


x=('t','u','g','d','k','l')
a,b,c,*d=x
print(x)  #('t', 'u', 'g', 'd')
print(a)    #t
print(b)    #u
print(c)    #g
print(d)    #d,k,l



#LOOP TUPLES

y=('a','k','w','t')
for i in y:
    print(i)
#a
#k
#l
#s
#q




for i in range(len(y)):
    print(y[i])
#a
#k
#l
#s
#q    


i=0
while i <len(y):
    print(y[i])
    i+=1    
#a
#k
#l
#s
#q 


x=('a','k','w')    #('a', 'k', 'w', 'q', 'a', 'd')
y=('q','a','d')
q=x+y
print(q)
w=x*2  #copy
print(w)   #('a', 'k', 'w', 'a', 'k', 'w')

w=('a', 'k', 'w', 'q', 'a', 'd')
y=w.count('k')  #1    counts the element
print(y)


w=('a', 'k', 'w', 'q', 'a', 'd')  #3    #checks the index
z=w.index('q')
print(z)

```

###  *set*

```python
#set
#{} use
#unchangeable (can't use index)
#unordered
#unindexed
#duplicates are not allowed  (1 true both are same value)
```

```python
nset={"apple","ball","call","dog"}   #4
print(len(nset))
print(type(nset))   #set


#ACCES ITEMS
nset={"apple","ball","cat","dog"}
for i in nset:
    print(i)



#add items


nset.add("eat")   #{'ball', 'dog', 'apple', 'cat', 'eat'}
print(nset)


#concatination
nset={'ball', 'dog', 'apple', 'cat', 'eat'}    #{}'goat', 'hat', 'cat', 'ball', 'eat', 'frog', 'dog', 'apple'}
kset={'frog','goat','hat'}
nset.update(kset)
print(nset)

nset.remove("ball")   #{'goat', 'hat', 'cat', 'eat', 'frog', 'dog', 'apple'}
print(nset)
  
nset.discard("frog")  #{'goat', 'hat', 'cat', 'eat', 'dog', 'apple'}
print(nset)

nset.discard("zebra")  # when the value is not in their didnt raise an error
print(nset)

nset.pop()  #pops and random item
print(nset)

nset.clear()  #clears an set
print(nset)


del nset  #deletes an set
print(nset)




nset.remove("bike")    #when the value is not their raise an error
print(nset)
```

```python
#set union

set1={1,2,3,4,5}
set2={5,6,7,8,9}

t1=set1.union(set2)     #{1, 2, 3, 4, 5, 6, 7, 8, 9}
print(t1)

#a method
t1=set1|set2
print(t1)


#set intersection
set3={1, 2, 3, 4, 5,}  #5
set4={5, 6, 7, 8, 9}
t2=set3.intersection(set4)
print(t2)

#a method

t2=set3&set4
print(t2)


#set diffrence    #1,2,3,4

set6={1,2,3,4,5}
set5={5,6,7,8,9}
t3=set6.difference(set5)
print(t3)

#a method
t3=set6-set5
print(t3)


#set symetrical diffrence   #1,2,3,4,6,7,8,9
set1={1,2,3,4,5}
set2={5,6,7,8,9}
t4=set1.symmetric_difference(set2)
print(t4)

#a method

t4=set1^set2
print(t4)
```

```python
cdic={'name':'ash',    #ash
      'age':'21',
      'colour':'black'}
print(cdic['name'])
d=cdic.get('age')     #21
print(d)

f=cdic.values()    #values
print(f)

print(cdic.items())   #  to get all values and keys

#if statement to check the keys
adic={'name':'ash',    
      'age':'21',
      'colour':'black'}


if 'name' in adic:    #yes
    print('yes')
else:
    print('no')


#if statement to check the values

adic={'name':'ash',    
      'age':'21',
      'colour':'black'}

if 'name' in adic.values():  #no
    print('yes')
else:
    print('no')



#to change the items


sk={'name':'ash',     #age =27   
      'age':'21',
      'colour':'black'
      }
sk['age']=27
print(sk)


sk={'name':'ash',     #{'name': 'ash', 'age': '21', 'colour': 'black', 'bike': 'yes'}
      'age':'21',
      'colour':'black'
      }
sk['bike']='yes'
print(sk)


#update the dict
ok={'name': 'ash',
     'age': '21',
       'colour': 'black', 
       'bike': 'yes'}
ok.update({"year":2003})  #{'name': 'ash', 'age': '21', 'colour': 'black', 'bike': 'yes', 'year': 2003}
print(ok)


ok={'name': 'ash', 'age': '21', 'colour': 'black', 'bike': 'yes', 'year': 2003}
ok.update({"age":21})    #{'name': 'ash', 'age': 21, 'colour': 'black', 'bike': 'yes', 'year': 2003}
print(ok)

ok["age"]=20   #{'name': 'ash', 'age': 20, 'colour': 'black', 'bike': 'yes', 'year': 2003}
print(ok)


#remove item
krthi={'name': 'ash', 'age': 20, 'colour': 'black', 'bike': 'yes', 'year': 2003}

#pop
krthi.pop("year")   #{'name': 'ash', 'age': 20, 'colour': 'black', 'bike': 'yes'}
print(krthi)

#pop item  (removes the last item)

krthi.popitem()  #'name': 'ash', 'age': 20, 'colour': 'black'}
print(krthi)

#del (you can deleth the dict  )  syntex  del.dict[key]

#clear  (you can clear the dict by using this)  syntex: dict.clear()
```
